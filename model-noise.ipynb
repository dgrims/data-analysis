```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
loc = np.logspace(1,4,20)
scale = loc*0.0125
```

```python
lam = np.vstack([np.random.normal(l,s,size=50000) for l,s in zip(loc,scale)])
```

```python
x = np.random.poisson(lam)
```

```python
sns.distplot(x[4])
```

```python
std = x.std(axis=1)
mean = x.mean(axis=1)
```

```python
plt.plot(mean,std,'.-')
plt.plot(loc,np.sqrt(loc),'-')
plt.xscale('log')
plt.yscale('log')
```
