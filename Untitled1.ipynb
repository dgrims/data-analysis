# Amplitude Damping

Kraus operators for amplitude damping:

```python
def g(T1):
    return 1-np.exp(-1./T1)
def k0(T1):
    return np.matrix([[1,0],[0,np.exp(-1./T1/2)]])
def k1(T1):
    return np.matrix([[0,np.sqrt(1-np.exp(-1./T1))],[0,0]])
```

```python
def k0(g):
    return np.matrix([[1,0],[0,np.sqrt(1-g)]])
def k1(g):
    return np.matrix([[0,np.sqrt(g)],[0,0]])
```

```python
# quantum register
q = QuantumRegister(1, 'q')

# measurement circuit
c = ClassicalRegister(1, 'c')
meas = QuantumCircuit(q, c)
meas.barrier(q)
meas.measure(q,c)

# quantum circuits
list_circ = []
circ = QuantumCircuit(q)
circ.x(q[0])
for t in range(0,20):
    circ.iden(q[0])
    list_circ.append(circ+meas)
```

```python
list_circ[18].draw(output='mpl')
```

```python
T1=5
shots = 1000
# create noise model object
noise_model = NoiseModel()

# create quantum error object from Kraus operators k0 and k1
error = kraus_error([k0(g(T1)), k1(g(T1))])

# add the quantum error for the identity gate and qubit 0 to the noise model
noise_model.add_quantum_error(error, 'id', [0])

# the noise_model option requires the backend Aer instead of BasicAer
# simulate a list of quantum circuits using the custom noise model
job = execute(list_circ, backend_sim, shots=shots, noise_model=noise_model)
result = job.result()

p1=np.array([ result.get_counts(n)['1'] if ('1' in result.get_counts(n)) else 0 for n,circ in enumerate(list_circ)])
p1_stderr=np.sqrt(p1*(shots-p1)/shots)
```

# b) Simulated Phase damping

Define the Kraus operator set for phase damping.

```python
def k0_phase(g2):
    return np.matrix([[1,0],[0,np.exp(-g2)]])
def k1_phase(g2):
    return np.matrix([[0,0],[0,np.sqrt(1-np.exp(-2*g2))]])
```

```python
w=1
T = 50
# quantum circuits
list_circ = []
circ = QuantumCircuit(q)
circ.h(q[0])

for t in range(0,T):
    circ.iden(q[0])
    ramsey = QuantumCircuit(q)
    ramsey.u1(t*w,q[0])
    ramsey.h(q[0])
    list_circ.append(circ+ramsey+meas)
    
```

```python
list_circ[3].draw()
```

```python
T1=5
T2=7
g2 = 1./T2-1./2/T1

shots = 1000
# create noise model object
noise_model = NoiseModel()

# create quantum error object from Kraus operators k0 and k1
error = kraus_error([k0_phase(g2), k1_phase(g2)])

# add the quantum error for the identity gate and qubit 0 to the noise model
noise_model.add_quantum_error(error, 'id', [0])

# the noise_model option requires the backend Aer instead of BasicAer
# simulate a list of quantum circuits using the custom noise model
job = execute(list_circ, backend_sim, shots=shots, noise_model=noise_model)
result = job.result()

p1=np.array([ result.get_counts(n)['1'] if ('1' in result.get_counts(n)) else 0 for n,circ in enumerate(list_circ)])
p1_stderr=np.sqrt(p1*(shots-p1)/shots)
```

```python
def damped_cos (x,A,y0,T,w,phi):  
    return A*np.exp(-x/T)*np.cos(w*x+phi)+y0
```

```python
model = Model(damped_cos)
params=model.make_params()
```

```python
params.add('y0',value=shots/2)
params.add('T',value=1./g2)
params.add('w',value=w)
params.add('phi',value=pi)
params.add('A',value=shots/2)
```

```python
x=np.arange(0,len(p1),1)
```

```python
fit_res = model.fit(p1,x=x,params=params,)
plt.figure(figsize=(20,10))
X = np.linspace(x.min(),x.max(),200)
plt.plot(X,model.eval(x=X,params=fit_res.params),label="Best fit")
#plt.plot(x,fit_res.init_fit)
plt.errorbar(x=x,y=p1,fmt=".",yerr=p1_stderr,label="'1' Count")
plt.xlabel('time in units of gate times')
plt.ylabel("counts for result '1'/'+'")
plt.legend()
```
